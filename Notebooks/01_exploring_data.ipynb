```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
species_df = pd.read_csv("../species.csv", sep=",")
```

```python
species_df
```

```python
plt.figure(figsize=(12, 12))
plt.title("Species Distribution")

sns.countplot(
    y="Common Name",
    data=species_df,
    palette="Set1",
    order=species_df["Common Name"].value_counts().index
)

plt.xlabel("Count")
plt.ylabel("Common Name")
plt.tight_layout()
plt.show()
```

```python
species_df["Scientific Name"].nunique()
```

```python
species_counts = species_df['Species eBird Code'].value_counts()
print(species_counts)

species_counts.plot(kind='bar', figsize=(14, 5))
plt.title('Number of annotations per species')
plt.ylabel('Count')
plt.xlabel('Species')
plt.tight_layout()
plt.show()
```

```python
annotation_df = pd.read_csv("../annotations.csv", sep=",")
```

```python
species_counts = annotation_df['Species eBird Code'].value_counts()
print(species_counts)

species_counts.plot(kind='bar', figsize=(14, 5))
plt.title('Number of annotations per species')
plt.ylabel('Count')
plt.xlabel('Species')
plt.tight_layout()
plt.show()
```

```python
species_counts = annotation_df['Species eBird Code'].value_counts()

for threshold in [500, 200, 100, 50, 20]:
    n_species = (species_counts >= threshold).sum()
    print(f'Species with >= {threshold} annotations: {n_species}')
```
